## 1. Data Visualization

### 1.1. ASD vs TDC per sites

```python
import pandas as pd
import numpy as np
import ast

node_based_features = pd.read_csv('./outputs/node_based_features.csv')
for col in [f"{atlas}_degree" for atlas in ['cc200', 'aal', 'dos160']]:
    node_based_features[col] = node_based_features[col].apply(ast.literal_eval)
```

```python
node_based_features.head()
```

```python
node_degrees = node_based_features[['id', 'ASD', 'SITE_ID']+[f"{atlas}_degree" for atlas in ['cc200', 'aal', 'dos160']]]
node_degrees.head()
```

Remove `_number` suffix from `SITE_ID`. For example, we have `UCLA_1` and `UCLA_2`. I will change the names to `UCLA` as we have same name in other benchmark researches.

```python
node_degrees.loc[:, 'SITE_ID'] = node_degrees['SITE_ID'].str.replace(r'_\d+$', '', regex=True)
# Count unique SITE_IDs after merging
num_sites = node_degrees['SITE_ID'].nunique()
print("Number of unique SITE_IDs after merging:", num_sites)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example structure of node_degrees dataframe
# node_degrees = pd.DataFrame({
#     'id': [1, 2, 3, 4, 5],
#     'ASD': [1, 0, 1, 0, 1],
#     'SITE_ID': ['Site1', 'Site1', 'Site2', 'Site2', 'Site1']
# })

# Group data by SITE_ID and ASD, then count occurrences
site_counts = node_degrees.groupby(['SITE_ID', 'ASD']).size().unstack(fill_value=0)
site_counts.columns = ['TDC', 'ASD']

# Calculate percentages
site_counts_percent = site_counts.div(site_counts.sum(axis=1), axis=0) * 100

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))
x = range(len(site_counts))  # X positions for each site

# Plotting each category (TDC and ASD) with a slight horizontal offset
width = 0.35  # Width of each bar
ax.bar([pos - width / 2 for pos in x], site_counts['TDC'], width=width, color='skyblue', label='TDC')
ax.bar([pos + width / 2 for pos in x], site_counts['ASD'], width=width, color='salmon', label='ASD')

# Adding vertical percentages on the bars
for idx, (site, counts) in enumerate(site_counts.iterrows()):
    for i, (label, count) in enumerate(counts.items()):
        percent = site_counts_percent.loc[site, label]
        x_pos = idx - width / 2 if label == 'TDC' else idx + width / 2
        ax.text(x_pos, count / 2, f"{percent:.1f}%", ha='center', va='center', rotation=90, fontsize=10, color='black')

# Adding titles and labels
plt.title("ASD vs TDC Counts per Site with Percentages")
plt.xlabel("Site", fontsize=14)
plt.ylabel("Count of Subjects", fontsize=14)
plt.xticks(x, site_counts.index, rotation=45)  # Set x-ticks to site names
plt.legend(title='Diagnosis')

plt.tight_layout()
plt.show()
```

### 1.2. Male vs Female per sites

```python
# Read the extra information from the CSV file
df_labels = pd.read_csv('pheno_file.csv')

# Merge `node_degrees` with `df_labels` on `id` and `subject`
node_degrees = node_degrees.merge(df_labels[['subject', 'SEX', 'AGE_AT_SCAN']], left_on='id', right_on='subject', how='left')

# Drop the redundant 'subject' column if needed
node_degrees = node_degrees.drop(columns=['subject'])

node_degrees.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example structure of node_degrees dataframe
# Assuming node_degrees already contains columns 'id', 'SEX', and 'SITE_ID'

# Group data by SITE_ID and SEX, then count occurrences
sex_counts = node_degrees.groupby(['SITE_ID', 'SEX']).size().unstack(fill_value=0)

# Calculate percentages
sex_counts_percent = sex_counts.div(sex_counts.sum(axis=1), axis=0) * 100

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))
x = range(len(sex_counts))  # X positions for each site

# Plotting each category (Male and Female) with a slight horizontal offset
width = 0.35  # Width of each bar
ax.bar([pos - width / 2 for pos in x], sex_counts_percent[1], width=width, color='lightblue', label='Male')
ax.bar([pos + width / 2 for pos in x], sex_counts_percent[2], width=width, color='lightcoral', label='Female')

# Adding percentages on the bars
for idx, (site, counts) in enumerate(sex_counts.iterrows()):
    for i, (label, count) in enumerate(counts.items()):
        percent = sex_counts_percent.loc[site, label]
        x_pos = idx - width / 2 if label == 1 else idx + width / 2
        if percent > 10:
            ax.text(x_pos, percent / 2, f"{percent:.1f}%", ha='center', va='center', rotation=90, fontsize=10, color='black')
        elif percent > 0: # Only add the text if count is greater than zero and for small count lower the fontsize to fit.
            ax.text(x_pos, percent / 2, f"{percent:.1f}%", ha='center', va='center', rotation=90, fontsize=8, color='black')

# Adding titles and labels
plt.title("Percentage of Male and Female Subjects per Site")
plt.xlabel("Site", fontsize=14)
plt.ylabel("Percentage (%)", fontsize=14)
plt.xticks(x, sex_counts.index, rotation=45)  # Set x-ticks to site names
plt.legend(title='Sex', labels=['Male', 'Female'])

plt.tight_layout()
plt.show()
```

## Feature Engineering

### Split the Dataset

First, Split the dataset into Train, Test, Validation

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Create a new column 'SITE_ID_MERGED' by removing the postfix from 'SITE_ID'
node_degrees['SITE_ID_MERGED'] = node_degrees['SITE_ID'].apply(lambda x: x.split('_')[0])

# Create empty lists to store the split data
train_list = []
val_list = []
test_list = []

# Group by 'SITE_ID_MERGED' and split the data for each site
for site, group in node_degrees.groupby('SITE_ID_MERGED'):
    # First split the group into train (80%) and temp (20%)
    train_data, temp_data = train_test_split(group, test_size=0.2, random_state=42, stratify=group['ASD'])
    
    # Now split the temp data into validation (50%) and test (50%) - effectively 10% of original for each
    val_data, test_data = train_test_split(temp_data, test_size=0.5, random_state=42, stratify=temp_data['ASD'])
    
    # Append the splits to their respective lists
    train_list.append(train_data)
    val_list.append(val_data)
    test_list.append(test_data)

# Concatenate all the splits to get the final train, val, and test datasets
train_data = pd.concat(train_list).reset_index(drop=True)
val_data = pd.concat(val_list).reset_index(drop=True)
test_data = pd.concat(test_list).reset_index(drop=True)

# Plotting the stacked bar plot
site_ids = node_degrees['SITE_ID_MERGED'].unique()
site_counts = node_degrees['SITE_ID_MERGED'].value_counts().sort_values(ascending=False)

# Sort the site IDs by the total number of samples for each site
site_ids = site_counts.index

train_counts = [len(train_data[train_data['SITE_ID_MERGED'] == site]) for site in site_ids]
val_counts = [len(val_data[val_data['SITE_ID_MERGED'] == site]) for site in site_ids]
test_counts = [len(test_data[test_data['SITE_ID_MERGED'] == site]) for site in site_ids]

bar_width = 0.5
x = range(len(site_ids))

plt.figure(figsize=(10, 6))

# Plotting stacked bars
plt.bar(x, train_counts, width=bar_width, color='blue', label='Train')
plt.bar(x, val_counts, width=bar_width, color='orange', bottom=train_counts, label='Validation')
plt.bar(x, test_counts, width=bar_width, color='red', bottom=[i+j for i,j in zip(train_counts, val_counts)], label='Test')

plt.xlabel('Site ID')
plt.ylabel('Number of Samples')
plt.title('Number of Train, Validation, and Test Samples per Site (Stacked)')
plt.xticks(x, site_ids, rotation=45)  # Rotate x-axis labels by 45 degrees
plt.legend()
plt.tight_layout()

plt.show()
```

```python
train_data['all_atlases_degree'] = train_data[['cc200_degree', 'aal_degree', 'dos160_degree']].apply(lambda x: np.concatenate(x), axis=1)
val_data['all_atlases_degree'] = val_data[['cc200_degree', 'aal_degree', 'dos160_degree']].apply(lambda x: np.concatenate(x), axis=1)
test_data['all_atlases_degree'] = test_data[['cc200_degree', 'aal_degree', 'dos160_degree']].apply(lambda x: np.concatenate(x), axis=1)
```

### Feature Selection

#### __1.__ Mutual Information

```python
import numpy as np
from sklearn.feature_selection import mutual_info_classif

# Create a dictionary to store the top feature indices for each SITE_ID_MERGED
top_feature_indices_dict = {}

# Group by SITE_ID_MERGED
for site_id, group in train_data.groupby('SITE_ID_MERGED'):
    # Extract the features and labels for the current SITE_ID_MERGED
    X_train_array = np.array([row for row in group['all_atlases_degree']])
    y_train = group['ASD'].values
    
    # Apply mutual information on the training data
    mutual_info = mutual_info_classif(X_train_array, y_train, n_neighbors=10, random_state=42)
    
    # Select the top features based on mutual information
    top_n = 50
    top_feature_indices = np.argsort(mutual_info)[::-1][:top_n]
    
    # Store the top feature indices in the dictionary
    top_feature_indices_dict[site_id] = top_feature_indices
    
    print(f"SITE_ID_MERGED: {site_id}, Top Feature Indices: {top_feature_indices}")
```

```python
site_id = 'UCLA'
top_feature_indices = top_feature_indices_dict[site_id]
print(f"Top Feature Indices for {site_id}: {top_feature_indices}")
```

#### __2.__ PCA

```python
import matplotlib.pyplot as plt

# Initialize a counter for each feature
feature_counter = np.zeros(477)

# Iterate over the top feature indices for each site
for site_id, top_feature_indices in top_feature_indices_dict.items():
    # Increment the counter for each feature that is in the top 50
    feature_counter[top_feature_indices] += 1

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(range(477), feature_counter)
plt.xlabel('Feature Index')
plt.ylabel('Number of Sites in Top 50')
plt.title('Feature Selection Frequency Across Sites')
plt.show()
```

```python
top_n_features = 20
top_features = np.argsort(feature_counter)[::-1][:top_n_features]
plt.bar(range(top_n_features), feature_counter[top_features])
plt.xlabel('Top Features')
plt.ylabel('Number of Sites in Top 50')
plt.title('Top Features Across Sites')
plt.show()
```
